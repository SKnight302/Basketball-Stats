```python
place["Team"]=place["Team"]+ " "+"(" + seed_pm["Conference"]+")"
place=place.drop(columns=["Conference"])
```

```python
merged_stats= pd.read_csv("./2012_csv_files/NCAA_Merged_Statistics_2011-12.csv")
full_stats = merged_stats.merge(place,how= 'inner')

place[~place['Team'].isin(full_stats['Team'])]
```

```python
all_stats= merged_stats.merge(place,how= 'outer')
all_stats = all_stats.fillna('DNP')

all_stats.to_csv('./2012_csv_files/NCAA_2011-12_All_Stats.csv')
```
