```python
import pandas as pd
seed_pm = pd.read_csv("./2015_csv_files/NCAA_Stats_2014-15_Placement.csv",encoding='cp1252')
seed_pm.head()
```

```python
place = seed_pm
place['Conference']= seed_pm['Conference'].replace('American','AAC')
place['Conference']= seed_pm['Conference'].replace('Missouri Valley','MVC')
place['Conference']= seed_pm['Conference'].replace('Ohio Valley','OVC')
place['Conference']= seed_pm['Conference'].replace('Atlantic Sun','ASUN')
place['Conference']= seed_pm['Conference'].replace('Conference USA','C-USA')
place['Conference']= seed_pm['Conference'].replace('Southern','SoCon')
place['Conference']= seed_pm['Conference'].replace('West Coast','WCC')
place['Conference']= seed_pm['Conference'].replace('Colonial','CAA')
place['Conference']= seed_pm['Conference'].replace('Ivy','Ivy League')
place['Conference']= seed_pm['Conference'].replace('Summit','Summit League')
place['Conference']= seed_pm['Conference'].replace('Mid American','MAC')

place['Team']= seed_pm['Team'].replace('State','St.',regex = True)
place['Team']= seed_pm['Team'].replace('NC St.','NC State',regex = True)
place['Team']= seed_pm['Team'].replace('North Carolina St.','NC State')
place['Team']= seed_pm['Team'].replace('Stephen F. Austin','SFA')
place['Team']= seed_pm['Team'].replace('Loyola–Chicago','Loyola Chicago')
place['Team']= seed_pm['Team'].replace('LIU Brooklyn','LIU')
place['Team']= seed_pm['Team'].replace('North Carolina Central','N.C. Central')
place['Team']= seed_pm['Team'].replace('College of Charleston', 'Col. of Charleston')
place['Team']= seed_pm['Team'].replace('USC','Southern California')
place['Team']= seed_pm['Team'].replace('Middle Tennessee','Middle Tenn.')
place['Team']= seed_pm['Team'].replace('East Tennessee St.','ETSU')
place['Team']= seed_pm['Team'].replace('UNC Wilmington','UNCW')
place['Team']= seed_pm['Team'].replace("Saint Mary's","Saint Mary's (CA)")
place['Team']= seed_pm['Team'].replace("Florida Gulf Coast","FGCU")
place['Team']= seed_pm['Team'].replace("Northern Kentucky","Northern Ky.")
place['Team']= seed_pm['Team'].replace("Miami","Miami (FL)")
place['Team']= seed_pm['Team'].replace("Northern Iowa","UNI")
place['Team']= seed_pm['Team'].replace("Cal St. Bakersfield","CSU Bakersfield")
place['Team']= seed_pm['Team'].replace("Southern","Southern U.")
place['Team']= seed_pm['Team'].replace("Mississippi","Ole Miss")
place['Team']= seed_pm['Team'].replace("Albany","Albany (NY)")
place['Team']= seed_pm['Team'].replace("St. John's","St. John's (NY)")
place['Team']= seed_pm['Team'].replace('Eastern Washington','Eastern Wash.')





place.head()
```

```python
place["Team"]=place["Team"]+ " "+"(" + seed_pm["Conference"]+")"
place=place.drop(columns=["Conference"])
```

```python
merged_stats= pd.read_csv("./2015_csv_files/NCAA_Merged_Statistics_2014-15.csv")
full_stats = merged_stats.merge(place,how= 'inner')
place[~place['Team'].isin(full_stats['Team'])]
```

```python
all_stats= merged_stats.merge(place,how= 'outer')
all_stats = all_stats.fillna('DNP')

all_stats.to_csv('./2015_csv_files/NCAA_2014-15_All_Stats.csv')
```

```python
complete_stats = all_stats[['Year','Team','Power 5 School','Placement','Seed','Berth type','Overall rank', 'Win Pct', 'SCR MAR', 'REB MAR', 'FG%','FT%', '3FG%', 'OPP FG%', 'OPP 3FG%','AST/TO Ratio']]
complete_stats.to_csv('./2015_csv_files/NCAA_2014-15_Completed_Stats.csv',index=False)
```
