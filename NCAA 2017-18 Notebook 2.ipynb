```python
import pandas as pd 
seed_pm = pd.read_csv("./2018_csv_files/NCAA_Stats_2017-18_Placement.csv",encoding='cp1252')
seed_pm.head()
```

```python
place = seed_pm
place['Conference']= seed_pm['Conference'].replace('American','AAC')
place['Conference']= seed_pm['Conference'].replace('Missouri Valley','MVC')
place['Conference']= seed_pm['Conference'].replace('Ohio Valley','OVC')
place['Conference']= seed_pm['Conference'].replace('Atlantic Sun','ASUN')
place['Conference']= seed_pm['Conference'].replace('Conference USA','C-USA')
place['Conference']= seed_pm['Conference'].replace('Southern','SoCon')
place['Team']= seed_pm['Team'].replace('State','St.',regex = True)
place['Team']= seed_pm['Team'].replace('NC St.','NC State',regex = True)
place['Team']= seed_pm['Team'].replace('Stephen F. Austin','SFA')
place['Team']= seed_pm['Team'].replace('Loyola–Chicago','Loyola Chicago')
place['Team']= seed_pm['Team'].replace('LIU Brooklyn','LIU')
place['Team']= seed_pm['Team'].replace('North Carolina Central','N.C. Central')
place['Team']= seed_pm['Team'].replace('College of Charleston', 'Col. of Charleston')
place.head(68)
```

```python
place["Team"]=place["Team"]+ " "+"(" + seed_pm["Conference"]+")"
place=place.drop(columns=["Conference"])
```

```python
merged_stats= pd.read_csv("./2018_csv_files/NCAA_Merged_Statistics_2017-18.csv")
full_stats = merged_stats.merge(place,how= 'inner')
place[~place['Team'].isin(full_stats['Team'])]
```

```python
all_stats= merged_stats.merge(place,how= 'outer')
all_stats = all_stats.fillna('DNP')
all_stats.to_csv('./2018_csv_files/NCAA_2017-18_All_Stats.csv')
```

```python
complete_stats = all_stats[['Year','Team','Power 5 School','Placement','Seed','Berth type','Overall rank', 'Win Pct', 'SCR MAR', 'REB MAR', 'FG%','FT%', '3FG%', 'OPP FG%', 'OPP 3FG%','AST/TO Ratio']]
complete_stats.to_csv('./2018_csv_files/NCAA_2017-18_Completed_Stats.csv')
```
