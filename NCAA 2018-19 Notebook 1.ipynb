```python
import pandas as pd
df = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_Won_Lost_Pct.csv")

df1 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_Scoring_Margin.csv")

df2 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_Rebound_Margin.csv")

df3 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_FG_Pct.csv")

df4 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_FT_Pct.csv")

df5 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_3FG_Pct.csv")

df6 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_OPP_FG_Pct.csv")

df7 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_OPP_3FG_Pct.csv")
df7.columns= ['Team','Opp 3FGAPG', 'Opp 3FGPG','Opp 3FG%']

df8 = pd.read_csv("./2019 csv_files/NCAA_Stats_2018-19_AST_TO_Ratio.csv")


merged_stats = df.merge(df1.merge(df2.merge(df3.merge(df4.merge(df5.merge(df6.merge(df7.merge(df8))))))))
merged_stats.head()
```

```python
merged_stats.to_csv("./2019 csv_files/NCAA_Merged_Statistics_2018-19.csv",index = False)
```

```python
merged_stats.insert(0,"Year",2019)
merged_stats.head()
```
