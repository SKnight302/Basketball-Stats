```python
import pandas as pd
df = pd.read_csv("./csv_files/NCAA Statistics 2020-21 Won-Lost Pct.csv")
df_dropped = df.drop(['Rank'],axis =1)
```

```python
df1 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 Scoring Margin.csv")
df1_dropped = df1.drop(columns =['Rank'])

df2 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 Rebound Margin.csv")
df2_dropped = df2.drop(columns =['Rank'])

df3 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 FG Pct.csv")
df3_dropped = df3.drop(columns =['Rank'])

df4 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 FT Pct.csv")
df4_dropped = df4.drop(columns =['Rank'])

df5 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 3-Point FG Pct.csv")
df5_dropped = df5.drop(columns =['Rank'])

df6 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 Opp FG pct.csv")
df6_dropped = df6.drop(columns =['Rank'])

df7 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 AST TO Ratio.csv")
df7_dropped = df7.drop(columns =['Rank'])

df8 = pd.read_csv("./csv_files/NCAA Statistics 2020-21 OPP 3-Point FG Pct.csv")
df8_dropped = df8.drop(columns =['Rank'])

merged_stats = df_dropped.merge(df1_dropped.merge(df2_dropped.merge(df3_dropped.merge(df4_dropped.merge(df5_dropped.merge(df6_dropped.merge(df7_dropped.merge(df8_dropped))))))))
merged_stats.head(25)
```

```python
merged_stats.insert(0,"Year",2021)
merged_stats.insert(2,"Power 5 School","No")
merged_stats.loc[(merged_stats['Team'].str.contains('Big 12|SEC|ACC|Big Ten|Pac-12', regex = True)),'Power 5 School'] = 'Yes'
```

```python
merged_stats.head()
```

```python
easy_stats = merged_stats[['Year','Team','Power 5 School', 'Win Pct', 'SCR MAR', 'REB MAR', 'FG%','FT%', '3FG%', 'OPP FG%', 'OPP 3FG%','AST/TO Ratio']]
easy_stats.head()
```

```python
merged_stats.to_csv("NCAA Merged P5 Statistics 2020-21.csv",index = False)
easy_stats.to_csv("NCAA Condensed Statistics 2020-21.csv",index = False)
```
