```python
import pandas as pd
merged_stats = pd.read_csv("NCAA Merged Statistics 2020-21.csv")
easy_stats = pd.read_csv("NCAA Condensed Statistics 2020-21.csv")
seed_pm = pd.read_csv("./csv_files/NCAA Statistics 2020-21 Seeding_Placement.csv")
```

```python
seed_pm.head()
```

```python
seed_pm["Team"]=seed_pm["Team"]+ " "+"(" + seed_pm["Conference"]+")"
seed_pm=seed_pm.drop(columns=["Conference"])
```

```python
full_stats = merged_stats.merge(seed_pm,how= 'outer')
full_stats = full_stats.fillna('DNP')
full_stats
```

```python
full_stats.insert(0,"Year",2021)
full_stats.insert(2,"Power 5 School","False")
full_stats.loc[(full_stats['Team'].str.contains('Big 12|SEC|ACC|Big Ten|Pac-12', regex = True)),'Power 5 School'] = 'True'
full_stats.head()
```

```python
full_stats.to_csv("./csv_files/NCAA_2020-21_All_Stats.csv")
```

```python
comp_stats = full_stats[['Year','Team','Power 5 School','Placement','Seed','Berth type','Overall rank', 'Win Pct', 'SCR MAR', 'REB MAR', 'FG%','FT%', '3FG%', 'OPP FG%', 'OPP 3FG%','AST/TO Ratio']]
comp_stats.to_csv("./csv_files/NCAA 2020-21_Completed_Stats.csv",index=False)
```
